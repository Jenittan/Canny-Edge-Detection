```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('Screenshot 2025-11-01 132312.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
plt.imshow(img, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```python
img.shape
```

```python
edges = cv2.Canny(img, threshold1=100, threshold2=200)
```

```python
plt.imshow(edges, cmap='gray')
```

```python
blurred = cv2.GaussianBlur(img_rgb, (7,7),0)
plt.imshow(blurred);plt.axis('off');plt.title('Blurred Image'); plt.show()
```

```python
medium=np.median(img)
```

```python
medium
```

```python
low=int(max(0,0.5*medium))
high=int(max(255,1/3*medium))
```

```python
low
```

```python
high
```

```python
canny_edges = cv2.Canny(blurred, threshold1=low, threshold2=high)
```

```python
plt.imshow(canny_edges)
```
